```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("KAG_conversion_data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.rename(columns={'ad_id': 'Ad_id',
                            'xyz_campaign_id': 'Campaign_id',
                            'fb_campaign_id': 'FB_Campaign_id',
                            'age': 'Age',
                            'gender': 'Gender',
                            'interest': 'Interest'})
```

```python
age_split = data['Age'].str.split("-", expand=True)
data['Age_New'] = (age_split[0].astype(int) + age_split[1].astype(int)) / 2
```

```python
data.Interest.unique()
```

```python
data.Approved_Conversion.max()
```

```python
data.columns
```

```python
# lets assume value per conversion is $10
VALUE_PER_CONVERSION = 10
```

```python
data['CTR'] = data['Clicks'] / data['Impressions']
data['CPC'] = data['Spent'] / data['Clicks']
data['CPM'] = (data['Spent'] / data['Impressions']) * 1000
data['Conversion_Rate'] = data['Approved_Conversion'] / data['Clicks']
data['CPA'] = data['Spent'] / data['Approved_Conversion']

data['Revenue'] = data['Approved_Conversion'] * VALUE_PER_CONVERSION
data['Profit'] = data['Revenue'] - data['Spent']
data["ROI"] = (data['Profit'] / data['Spent']) * 100
```
